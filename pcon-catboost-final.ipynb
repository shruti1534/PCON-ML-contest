```python
train_path="/kaggle/input/pcon-ml-contest/train.csv"
train_data=pd.read_csv(train_path)
```

```python
test_path="/kaggle/input/pcon-ml-contest/test.csv"
test_data=pd.read_csv(test_path)
```

```python
y=train_data.y
```

```python
featured_data = ['age', 'job', 'marital', 'education', 'default', 'housing', 'loan',
       'contact', 'month', 'day_of_week', 'duration', 'campaign', 'pdays',
       'previous', 'poutcome', 'emp.var.rate', 'cons.price.idx',
       'cons.conf.idx', 'euribor3m', 'nr.employed', 'ID']
X=train_data[featured_data]
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
categorical_values = X.select_dtypes(include = object).columns
```

```python
for x in categorical_values:
    X[x] = label_encoder.fit_transform(X[x])
```

```python
from sklearn.metrics import f1_score
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, random_state=1)
```

```python
from catboost import CatBoostClassifier
model = CatBoostClassifier(
    iterations=1200,
    learning_rate=0.015,
    random_seed=0,
    early_stopping_rounds=20
)
model.fit(
    X_train, y_train,
    verbose=False,
    eval_set=(X_valid,y_valid),
    plot=True
)
```

```python
y_pred=model.predict(X_valid)
```

```python
f1_score(y_valid,y_pred,pos_label='yes')
```

```python
test_X = test_data[featured_data]
for t in categorical_values:
    test_X[t] = label_encoder.fit_transform(test_X[t])
```

```python
prediction = model.predict(test_X)
prediction
```

```python
Id = test_data.ID
```

```python
my_submission = pd.DataFrame({'Id':Id , 'Predicted':prediction})
```

```python
my_submission.to_csv('Submissn.csv', index=False)
```
